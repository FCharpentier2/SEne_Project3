### Import datas

```python
df_france_raw=pd.read_excel('bilan_energetique_france_2021_series_longues.xlsx')
df_france_raw.head()
```

### Cleaning datas

```python
## Slicing on metropolitan France (more datas)

df_france_raw=df_france_raw[df_france_raw['NIVGEO']=='Métropole']
df_france_raw.shape
```

```python
## Dropping years 1960 - 2010 (not enough datas)
df_france_raw.drop(df_france_raw.columns[7:58],axis=1, inplace=True)
```

```python
## Grouping the source of energy

df_france_clean=df_france_raw.groupby(['TYPFLUX','ENERGIE'], as_index=False).sum()
```
